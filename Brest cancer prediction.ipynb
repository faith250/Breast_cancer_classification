```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neural_network import MLPClassifier
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
```

```python
def evaluate_model(model, X_test, y_test):
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    auc = roc_auc_score(y_test, y_pred)
    print(f'Accuracy: {accuracy:.4f}')
    print(f'Precision: {precision:.4f}')
    print(f'Recall: {recall:.4f}')
    print(f'F1 Score: {f1:.4f}')
    print(f'ROC AUC: {auc:.4f}')
```

```python
lr = LogisticRegression()
lr.fit(x_train, y_train)
print("Logistic Regression:")
evaluate_model(lr, x_test, y_test)
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train, y_train)
print("K-Nearest Neighbors:")
evaluate_model(knn, x_test, y_test)
```

```python
svm = SVC(kernel='linear', probability=True)
svm.fit(x_train, y_train)
print("Support Vector Machine:")
evaluate_model(svm, x_test, y_test)
```

```python
dt = DecisionTreeClassifier(random_state=42)
dt.fit(x_train, y_train)
print("Decision Tree:")
evaluate_model(dt, x_test, y_test)
```

```python
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(x_train, y_train)
print("Random Forest:")
evaluate_model(rf, X_test, y_test)
```

```python
gb = GradientBoostingClassifier(n_estimators=100, learning_rate=0.1, random_state=42)
gb.fit(x_train, y_train)
print("Gradient Boosting:")
evaluate_model(gb, x_test, y_test)
```

```python
nb = GaussianNB()
nb.fit(x_train, y_train)
print("Naive Bayes:")
evaluate_model(nb, x_test, y_test)
```

```python
ann = MLPClassifier(hidden_layer_sizes=(64, 32), max_iter=500, random_state=42)
ann.fit(x_train, y_train)
print("Artificial Neural Network:")
evaluate_model(ann, x_test, y_test)
```

```python
qda = QuadraticDiscriminantAnalysis()
qda.fit(x_train, y_train)
print("Quadratic Discriminant Analysis:")
evaluate_model(qda,x_test, y_test)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Dictionary to store metrics
metrics = {
    "Model": [],
    "Accuracy": [],
    "Precision": [],
    "Recall": [],
    "F1 Score": [],
    "ROC AUC": []
}

# Function to evaluate model and store results
def evaluate_and_store(model_name, model, x_test, y_test):
    y_pred = model.predict(x_test)
    metrics["Model"].append(model_name)
    metrics["Accuracy"].append(accuracy_score(y_test, y_pred))
    metrics["Precision"].append(precision_score(y_test, y_pred))
    metrics["Recall"].append(recall_score(y_test, y_pred))
    metrics["F1 Score"].append(f1_score(y_test, y_pred))
    metrics["ROC AUC"].append(roc_auc_score(y_test, y_pred))

# List of models to iterate through
models = [
    ("Logistic Regression", LogisticRegression()),
    ("K-Nearest Neighbors", KNeighborsClassifier(n_neighbors=5)),
    ("Support Vector Machine", SVC(kernel='linear', probability=True)),
    ("Decision Tree", DecisionTreeClassifier(random_state=42)),
    ("Random Forest", RandomForestClassifier(n_estimators=100, random_state=42)),
    ("Gradient Boosting", GradientBoostingClassifier(n_estimators=100, learning_rate=0.1, random_state=42)),
    ("Naive Bayes", GaussianNB()),
    ("Artificial Neural Network", MLPClassifier(hidden_layer_sizes=(64, 32), max_iter=500, random_state=42)),
    ("Quadratic Discriminant Analysis", QuadraticDiscriminantAnalysis())
]

# Fit each model and store metrics
for model_name, model in models:
    model.fit(x_train, y_train)
    evaluate_and_store(model_name, model, x_test, y_test)

# Convert metrics to DataFrame
metrics_df = pd.DataFrame(metrics)
print(metrics_df)
```

```python
# Set up a larger figure
plt.figure(figsize=(12, 8))

# Plot each metric in a separate subplot
metrics_to_plot = ["Accuracy", "Precision", "Recall", "F1 Score", "ROC AUC"]
for i, metric in enumerate(metrics_to_plot, 1):
    plt.subplot(2, 3, i)
    sns.barplot(x="Model", y=metric, data=metrics_df)
    plt.xticks(rotation=45)
    plt.title(f"Model Comparison by {metric}")

# Adjust layout and show plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Define a color palette for better distinction
palette = sns.color_palette("tab10", len(metrics_df['Model']))

# Set up the figure size
plt.figure(figsize=(16, 12))

# Use metrics to plot and define colors
metrics_to_plot = ["Accuracy", "Precision", "Recall", "F1 Score", "ROC AUC"]

for i, metric in enumerate(metrics_to_plot, 1):
    plt.subplot(3, 2, i)
    
    # Bar plot
    sns.barplot(x="Model", y=metric, data=metrics_df, palette=palette)
    plt.xticks(rotation=45)
    plt.title(f"Bar Plot - Model Comparison by {metric}")
    plt.xlabel("")

    # Line plot
    plt.subplot(3, 2, i + len(metrics_to_plot))
    sns.lineplot(x="Model", y=metric, data=metrics_df, marker="o", sort=False, color="teal")
    plt.xticks(rotation=45)
    plt.title(f"Line Plot - Model Comparison by {metric}")
    plt.xlabel("")

# Adjust layout and display plots
plt.tight_layout()
plt.show()
```
